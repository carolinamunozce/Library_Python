#Subsetting Columns

```python

# I can choise the columns what i want to work for a dataframe
df = df[['title', 'description', 'publishTime', 'kind_stats', 'duration_seconds',
       'viewCount', 'likeCount', 'commentCount']]
```

#Subsetting rows

```python
df[['viewCount']] > 1_000_000 # _ separator miles  it responds with true or false
```

```python
#it responds which one has the params search
#Subsetting using loc 
df_subset1 = df.loc[df['viewCount'] > 1_000_000 ]
```

```python
# It's the same when i use query on sql
df_subset2 = df.query('viewCount > 1000000')
```

```python
# Create on a dataframes all of values that are not 'NA'
df= df.loc[~df['viewCount'].isna()]
df= df.loc[~df['likeCount'].isna()]
# See when there are "nan" value, if i want to search the opossite, i have to use ~
```

# Casting dtypes

```python
df.dtypes
```

```python
#Changes the types 
df['viewCount'] =  df['viewCount'].astype('int')
df['likeCount'] =  df['likeCount'].astype('int')
```

```python
df.dtypes
```

```python
# Change datetime from object
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

```python
df['likeCount'] = pd.to_numeric(df['likeCount'].astype('str'))
```
