```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid') #conf de aspecto
```

```python
df = pd.read_csv('players_22.csv', low_memory=False) #Elimina la advertencia
```

```python
df = df[['short_name','age', 'nationality_name', 'overall', 'potential', 'club_name','value_eur','wage_eur','player_positions']]
```

```python
df['player_positions'].str.split(',', expand=True) #separa posiciones
```

```python
df['player_positions'] = df['player_positions'].str.split(',', expand=True)[0]
```

```python
df
```

```python
df.dropna(inplace=True)
df
```

```python
players_missing_worldcup = ['K. Benzema', 'S. Mané', 'S. Agüero', 'Sergio Ramos', 'P. Pogba',
                            'M. Reus', 'Diogo Jota', 'A. Harit', 'N. Kanté', 'G. Lo Celso', 'Piqué']
```

```python
eliminar_index = df[df['short_name'].isin(players_missing_worldcup)].index
```

```python
df.drop(eliminar_index, axis = 0, inplace= True) #inplace, guarda cambios , axis= eliminar filas, axis ouede ser filas o columnas por eso hay que decir cual se necesita
```

```python
#filtrar paises al mundial
df = df[df['nationality_name'].isin(teams_worldcup)]
```

##### 4. La mejor plantilla de cada país

```python
def mejor_plantilla(nacionalidad):
    df_best_squad = df.copy()
    df_best_squad = df_best_squad.groupby(['nationality_name','player_positions']).head(2) # escoger 2 jugadores por cada posicion
    df_best_squad = df_best_squad[df_best_squad['nationality_name'] == nacionalidad].sort_values(['player_positions', 'overall', 'potential'], ascending =False)
    return(df_best_squad)
```

```python
mejor_plantilla('Brazil')['overall'].mean()
```

```python
for team in teams_worldcup:
    print(team, mejor_plantilla(team)['overall'].mean())
```

```python
promedio_puntuacion = [mejor_plantilla(team)['overall'].mean() for team in teams_worldcup]
df_promedio_puntuacion = pd.DataFrame({'Teams': teams_worldcup, 'Promedio_Puntuacion': promedio_puntuacion})
df_promedio_puntuacion = df_promedio_puntuacion.dropna()
df_promedio_puntuacion = df_promedio_puntuacion.sort_values('Promedio_Puntuacion', ascending= False)
```

```python
df_promedio_puntuacion
```

```python
fig , ax = plt.subplots(figsize=(10, 6), tight_layout=True)

sns.barplot(data=df_promedio_puntuacion[:10], x='Teams', y='Promedio_Puntuacion')
plt.show()
#df_promedio_puntuacion
```
