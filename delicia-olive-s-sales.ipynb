```python
#What are the 5 best broughter?
```

```python
df.groupby('Clientes')['Total_Pago'].sum() \
    .sort_values(ascending=False) \
    .head(5)
```

```python
df.groupby('Clientes')['Total_Pago'].sum() \
    .sort_values(ascending=False) \
    .head(5) \
    .sort_values()\
    .plot(kind='barh',
         figsize=(5,4),
         title='Top 5 customer sales')
plt.show()
```

```python
df.query('Producto !="Pimenton y Ajo"') \
    .groupby('Clientes')['Total_Pago'].sum() \
    .sort_values(ascending=False) \
    .head(5)
```

```python
df.query('Producto != "Pimenton y Ajo"') \
   .groupby('Clientes')['Total_Pago'] \
   .agg(['mean','count']) \
   .query('mean>20000') \
   .sort_values('mean')['mean'] \
   .plot(kind='barh', figsize=(12,5), title='')
plt.show()
```
